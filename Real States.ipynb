## train test splitting


```python
# phele ek function banaya split train ka 
#fir random value li but seed se values fix ho gai matlab train aur test dono set ki 
#value fix ho gi 
#fir hum ne test set ka size fix kiya 
#fir test and train indices liye aur slicing apply ki 
# fir hum ne train_set aur test_set liye aur split function ko call kiya aur train set 70% and 
#test set 30% liya 
```

```python
import numpy as np

def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    print(shuffled)
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
train_set ,test_set = split_train_test(house,0.3)
```

```python


print(f"Rows in train set: {len(train_set)}\n Rows in test set :{len(test_set)}\n")
```

```python
#mainly it has three type of objects:
   # 1. esitmtors :it esitmates some parameters based on a dataset eg. imputer
  #  2. Transformers : it takes the inputs and result as output based on learning fits()
   # 3. predictors : linearRegresion function as predict
```

# create pipeline

```python
from sklearn.pipeline import  Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline =Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler())
])
```

```python
house_num=my_pipeline.fit_transform(house)
```

```python
house_num.shape
```

# selecting a desired model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(house_num,house_label)      #.....here hum ne house_label me 
```

```python
## now model get ready now check the prediction so apply some data through pipline
```

```python
some_data=house.iloc[:5]
```

```python
some_label=house_label.iloc[:5]
```

```python
prepared_data=my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_label)
```

```python
##evaluting model
```

```python
from sklearn.metrics import mean_squared_error
house_predicitons=model.predict(house_num)
lin_mse=mean_squared_error(house_label,house_predicitons)
lin_mse=np.sqrt(lin_mse)
```

```python
lin_mse
```

# cross vlidation technique

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(model,house_num,house_label,scoring="neg_mean_squared_error")
mse=np.sqrt(-score)
```
